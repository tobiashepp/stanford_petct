```python
import pathlib as plb
import pandas as pd
import shutil
```

```python
base_dir = plb.Path('/Users/gitijavidi/Box/Natasha-Sheybani-Files/NeroGCP_Identified_IRB43375')
root_dirs = [base_dir/'identified/dicom/for-alex-pt2',base_dir/'identified/dicom/Unzipped',
             base_dir/'identified-second-set/dicom/Unzipped',base_dir/'identified-second-set/dicom/for-alex-pt1']
```

```python
results = {'root': [], 'subject': [], 'final_0': [], 'final_1': []}
for r in root_dirs:
    print(f'parsing {r}')
    subject_dirs = [d for d in r.glob('SPC*') if d.is_dir()]
    for d in subject_dirs:
        final_dicoms = [f.name for f in (d/'Metavol/FinalDicoms').glob('*') if f.is_dir()]
        if len(final_dicoms) == 2:
            results['root'].append(str(r))
            results['subject'].append(d.name)
            results['final_0'].append(final_dicoms[0])
            results['final_1'].append(final_dicoms[1])
```

```python
df = pd.DataFrame(results)
df.head(5)
```

```python
df.to_csv('directory_list.csv')
```

## Extract data

```python
df = pd.read_csv('directory_list.csv')
df
```

```python
# unzipping all tgz which have a corresponding unzipped final dicoms directory
unzip = True 
tar_dir = plb.Path('/Users/gitijavidi/Desktop/pet_tars')
tar_list = list(tar_dir.glob('*.tgz'))
print('Extracting data from :', [t.name for t in tar_list])
for t in tar_list:
    subject = t.stem
    if subject in list(df['subject']):
        tar_path = t
        target_dir = tar_path.parent/tar_path.stem
        target_dir.mkdir(exist_ok=True)
        print(f'Unzipping {tar_path.name} to {target_dir} ...')
        if unzip:
            shutil.unpack_archive(tar_path, target_dir)
        dicom_dirs = list(next(target_dir.glob('*')).glob('*'))
        # Creating final dicoms directory for matching series
        subject_id = tar_path.stem
        select = df[df['subject']==subject_id]
        for series_id in dicom_dirs:
            if (select['final_0'].values[0] == series_id.name) or (select['final_1'].values[0] == series_id.name):
                print(f'Moving {series_id.name} to FinalDicoms')
                out_dir = series_id.parent.parent/'Metavol/FinalDicoms'
                out_dir.mkdir(parents=True, exist_ok=True)
                shutil.copytree(series_id, out_dir/series_id.name, dirs_exist_ok=True)
    else:
        print(f'No unzipped version for {t.name}')
```
